#Funciones para dividir

```python
def Inv(a,n):
  b=(a**(n-2))%n
  return b

def AlgDiv(f,g,n):
  #Obtenemos coeficientes
  Coef_Up=f.copy()
  Coef_Down=g.copy()
  #Obtenemos Grados
  GU=len(Coef_Up)
  GD=len(Coef_Down)
  #Obtenemos el inverso del primer elemento
  I=Inv(Coef_Down[0],n)
  #Guardaremos los coeficientes en este elemento
  Resul=[]
  for i in range(GU-(GD-1)):
    Resul.append(0)
  j=0
  #Mientras el residuo sea de grado mayor a g
  while(len(f)>=len(g)):
    Coef=f.copy()

    while(Coef[0]==0 and len(Coef)>1):
      Coef.pop(0)



    #Igualamos al grado en el que deberiamos estar, esto en caso de que no exista
    while (len(Coef)!=GU-j and GU-j!=0):
      Coef.insert(0,0)





    #Creamos un Auxiliar para cada Iteracion
    Aux=[]
    #Dividimos el coeficiente de arriba por el de abajo para obtener el coeficiente del resultado, agregamos al resultado
    Aux.append((Coef[0]*I)%n)
    Resul[j]=Aux[0]


    #Lo convertimos a polinomio, esto igualando el array a el grado que deberia ser,  esto mendiante
    # X^n/X^m=X^(n-m), solo que le restamos la iteracion puesto que es el grado siguinte
    Grad=GU-GD-j
    for i in range(Grad):
      Aux.append(0)
    #Coenvertimos a polinomio
    S=Aux.copy()
    #Se Multiplica el de abajo por el resultado de esta iteracion
    S=Mult(S,g,n)
    #Se Resta el resultado para la nueva iteracion
    f=Rest(f,S,n)
    #Obtenemos nueva iteracion
    j+=1

    while(f[0]==0 and len(f)>1):
      f.pop(0)
    if GU==1 or sum(f)==0:
      break


  #Retornamos
  return Resul,f

def Euc(f,g,n):
  orf=f.copy()
  org=g.copy()
  r=1
  C1=[[1],[0]]
  C2=[[0],[1]]
  ast=[]
  i=0

  print("\n\n Algoritmo de Euclides: ")
  #Mientras no tenga residuo 0
  while(numpy.sum(r)!=0):
    #Algoritmo de la division
    a,r=AlgDiv(f,g,n)
    #Guardamos resultado
    ast.append(a)

    #Tabulamos
    C1.append(Rest(C1[i],Mult(C1[i+1],ast[i],n),n))
    C2.append(Rest(C2[i],Mult(C2[i+1],ast[i],n),n))

    #Imprimimos recursion
    lector(f,n)
    print(" es igual a ",end=" ")
    lector(a,n)
    print(" multiplicado por ",end=" ")
    lector(g,n)
    print("  sumado con  ",end="")
    lector(r,n)
    print("")

    #Preparamos siguiente iteracion
    f=g
    g=r
    i+=1

  print("Aplicando inverso para hacer al polinomio monico tenemos que:")

  #Ponemos en forma a las tabulaciones no en forma:[0,0,0,2,1]->en forma [2,1]
  a1=Eliminador(C1[len(C1)-2])
  a2=Eliminador(C2[len(C2)-2])
  #Usamos las tabulaciones para obtener el gcd
  Comp=Sum(Mult(C1[len(C1)-2],orf,n),Mult(C2[len(C2)-2],org,n),n)
  Comp=Eliminador(Comp)

  #Conversion a Monico, se obtiene el inverso y todo se multiplica por el inverso
  Inver=Inv(Comp[0],n)
  a1=Esca(Inver,a1,n)
  a2=Esca(Inver,a2,n)
  Comp=Esca(Inver,Comp,n)
  f=Esca(Inver,f,n)

  #imprimimos resultados
  print("Maximo Comun Polinomio Divisor igual a ",end=" ")
  lector(f,n)


  print("\n\n\n-------------------------------------------------------------------------------------------------------------------------------------------------")
  #Cabe aclarar que los coeficientes de Bezout no son los divisores
  print("Es decir, que el maximo comun divisor se puede obtener con sus coeficientes de Bezout de la siguente manera:")
  lector(a1,n)
  print(" multiplicado por",end=" ")
  lector(orf,n)
  print("  sumado con  ",end=" ")
  lector(a2,n)
  print(" multiplicado por",end=" ")
  lector(org,n)
  print("  que es igual a:  ",end=" ")
  lector(Comp,n)
  print("\n\n\n")

  return f

def Eliminador(f):
  while(f[0]==0 and len(f)>1):
    f.pop(0)
  return f

def lector(f,n):
  Coef=f
  for i in range(len(Coef)-1):
    print(Coef[i]%n,"*x^",len(Coef)-i-1,sep="",end=" + ")
  print(Coef[len(Coef)-1],end="")
```

#Operaciones de Polinomios

```python
def Mult(f,g,n):
  s=numpy.convolve(f,g)
  for i in range(len(s)):
    s[i]=s[i]%n

  s=s.tolist()
  return s

def Esca(a,f,n):
  for i in range(len(f)):
    f[i]=a*f[i]%n
  return f

def Rest(f,g,n):
  if len(f)>len(g):
    maxi=len(f)
    mini=len(g)
    s1=f.copy()
    s2=g

    for i in range(mini):
      s1[maxi-i-1]=s1[maxi-i-1]-s2[mini-i-1]

  else:
    maxi=len(g)
    mini=len(f)
    s1=g.copy()
    s2=f
    for i in range(maxi):
      s1[i]=-s1[i]
    for i in range(mini):
      s1[maxi-i-1]=s1[maxi-i-1]+s2[mini-i-1]


  for i in range(len(s1)):
    s1[i]=s1[i]%n
  return s1

def Sum(f,g,n):
  if len(f)>len(g):
    maxi=len(f)
    mini=len(g)
    s1=f.copy()
    s2=g
  else:
    maxi=len(f)
    mini=len(g)
    s1=g.copy()
    s2=f
  for i in range(mini):
    s1[maxi-i-1]=s1[maxi-i-1]+s2[mini-i-1]
  for i in range(len(s1)):
    s1[i]=s1[i]%n
  return s1
```
