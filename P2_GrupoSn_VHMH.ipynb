#Funciones
Aqui encontrara las funciones que ocupa el programa para funcionar, se listan a continuacion:


*   Generador: Genera al grupo Sn
*   Ciclos: Obtiene los ciclos del elemento
*   Transposiciones: Calcula las transposiciones del elemento
*   Factorizacion: Encuentra la factorizacion a partir de las transposiciones del elemento
*   Recursion: Obtiene una factorizacion de una transposicion que no sea de la forma (i    i+1)
*   ImpresionFact: Lee el vector de factorizaci칩n y lo convierte en notacion simple

Nota: n siempre es el natural que se pidi칩 para generar al grupo


```python
#Generacion del grupo simetrico
def Generador(n):
  #Generamos la identidad
  Orig=np.arange(n)
  #Generamos todos sus elementos
  Per=np.array(list(it.permutations(Orig)))
  return Per


#Obtenemos los ciclos
def Ciclos(n, Obj):
  #Almacenamos los ciclos
  Cic=[]
  #Y los lugares que ya pasamos, como en un camino, para evitar repetir
  Pase=[]

  #Empezamos con la iteracion
  for i in range(n):
    #Variables auxiliares
    Aux=[]
    Aux2=i
    #Mientras no repitamos elementos seguiremos en el ciclo, cuando se repita alguno, entonces es decir que completamos el ciclo
    #Y pasaremos al siguiente, luego se verifica que no se encuentre en otro ciclo, asi aseguramos los minimos elementos que nos falten
    while Aux2 not in Aux and Aux2 not in Pase:
      #Cuando avancemos, decimos que ya pasamos por este elemento
      Pase.append(Aux2)
      #Lo agregamos al ciclo
      Aux.append(Aux2)
      #Pasamos al siguiente elemento
      Aux2=Obj[Aux2]
    #Finalmente, si el ciclo no es fixed (De un elemento) ni vacio, entonces se agrega
    if len(Aux)>1:
      Cic.append(Aux)
  #Retornamos ciclos
  return Cic

#Obtencion de Transposiciones Proposicion 2.35 Rotman
def Transposiciones(Obj):
  #Guardamos transposiciones
  Trans=[]
  #Obtenemos Transposiciones, para cada elemento de los ciclos
  for i in Obj:
    #Si ya es de tama침o 2 se agrega
    if len(i)==2:
      Trans.append(i)
    #Si no, entonces agregamos sus transposiciones, que son iguales a operar un ciclo de tama침o n de la siguiente manera:
    #(i_0,i_n)(i_0,i_(n-1))...(i_0,i_2)(i_0,i_1)
    else:
      for j in range(len(i)-1):
        Trans.append([i[0],i[len(i)-(1+j)]])
  #Retornamos Transposiciones
  return Trans

#Factorizacion
def Factorizacion(Obj):
  #Guardamos la Facotrizacion en forma de vectores [a,a+1], para esto usamos las transposiciones
  Fact=[]
  #Para cada elemento de la transposicion
  for i in Obj:
    #Medimos su distancia
    Dist=i[1]-i[0]
    #Si su distacia es de 1, entonces se agrega debido a que [i,i+1] puede ser representado
    if Dist==1:
      Fact.append(i)
    #Si no
    else:
      #Hacemos una recursion hasta que tengamos elementos de la forma [a,a+1]
      T=Recursion(i)
      #Agregamos a Fact
      for j in T:
        Fact.append(j)
  #Caso identidad
  if len(Fact)==0:
    Fact.append([-1,0])
  #Retornamos la factorizacion
  return Fact

#Devuelve una factorizacion en elementos de la forma [a,a+1]
def Recursion(Obj):
  #Checamos la distancia
  Dist=Obj[1]-Obj[0]
  #Caso base, acabando recursion si su distancia es 1
  if Dist==1:
    return Obj
  #Si no, recordemos que elementos de la forma [i,j] son iguales a [i,i+1][i+1,j][i+1,i]
  #Luego por como se trata la representacion tenemos que [i,i+1]=[i+1,i] puesto que i->i+1->i
  #Por lo que tenemos que es igual a [i,i+1][i+1,j][i,i+1]
  else:
    T=[]
    #Agregamos [i,i+1]
    T.append([Obj[0],Obj[0]+1])
    #Recursion hasta caso base de [j-1,j]
    #Retorna un vector [i,i+1][i+1,i+2]...[j-2,j-1][j-1,j][j-2,j-1]...[i+1,i+2][i,i+1] con i la cantidad de retornos hechos desde que llego al caso base
    Aux=Recursion([Obj[0]+1,Obj[1]])
    #Para agregar caso base
    if len(Aux)==2:
      T.append(Aux)
    #Para agregar casos no base
    else:
      for i in Aux:
        T.append(i)
    T.append([Obj[0],Obj[0]+1])
    return T

#Lectura de Factorizacion
def impresionFact(Obj):
  #Para cada elemento de Factorizacion
  for i in Obj:
    #Caso identidad
    if i[0]==-1:
      print("[]", end=" ")
    #Caso S=[0,1]
    elif i[0]==0:
      print("s",end=" ")
    #Caso [i,i+1]=r^isr^-i
    else:
      print("r^",i[0],"s","r^(-",i[0],")",sep="",end=" ")
  return
```
